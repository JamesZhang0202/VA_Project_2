```python
import pandas as pd
import numpy as np
```

```python
mobile = pd.read_excel(open('Mobile Device Data for Assignment 2.xlsx', 'rb'), sheet_name='Normalized Product Data', engine='openpyxl') 
```

```python
company =  pd.read_excel(open('Mobile Device Data for Assignment 2.xlsx', 'rb'), sheet_name='Model-Company', engine='openpyxl') 
```

```python
companyID = pd.read_excel(open('Mobile Device Data for Assignment 2.xlsx', 'rb'), sheet_name='Company ID', engine='openpyxl') 
```

```python
mobile.dropna(inplace = True)
mobile.reset_index(drop=True, inplace=True)
company.dropna(inplace = True)
company.reset_index(drop=True, inplace=True)
companyID.dropna(inplace = True)
companyID.reset_index(drop=True, inplace=True)
```

```python
company_companyID = pd.merge(company,companyID,how="inner",left_on="Company_ID",right_on="Company_ID")
company_companyID
```

```python
data =  pd.merge(mobile, company_companyID,how="outer",left_on="Model ID",right_on="Model ID")
data
```

```python
drop_list = ["Model_y", "Release Date_y", "Release Year_y"]
data = data.drop(columns=drop_list)
data.rename({'Model_x': 'Model', 'Release Date_x': 'Release Date', 'Release Year_x':'Release Year'}, axis=1, inplace=True)
```
